# Modeling - Simple manual model

```python
data["target_names"]
```

```python
def single_feature_prediction(petal_lenght):
    if petal_lenght <2.5:
        return 0
    elif petal_lenght < 4.8:
        return 1
    else:
        return 2 
```

```python
x_train[:, 2]
```

```python
manual_y_prediction= np.array([single_feature_prediction(val) for val in x_train [:,2]])
```

```python
manual_model_accuracy = np.mean(manual_y_prediction == y_train)
```

```python
print(f"Manual model accuracy: {manual_model_accuracy * 100:.2f}%")
```

# Modeling - Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
```

## Using a validation set to evaluate our model

```python
model = LogisticRegression()
```

```python
# xt = x_train and xv = x_validation
xt, xv, yt, yv = train_test_split(x_train, y_train, test_size=0.25)
```

```python
model.fit(xt, yt)
```

```python
y_pred = model.predict(xv)
```

```python
np.mean(y_pred == yv)
```

```python
model.score(xt, yt)
```

## using cross validation to evaluate our model

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
```

```python
np.mean(accuracies)
```

```python
y_pred = cross_val_predict(model, x_train, y_train, cv =5)
```

```python
predicted_correctly_mask = y_pred == y_train
```

```python
not_predicted_correctly = ~predicted_correctly_mask
```

```python
x_train[not_predicted_correctly]
```

```python
df_prediction = df_train.copy()
```

```python
df_prediction["correct_prediction"] = predicted_correctly_mask
```

```python
df_prediction["prediction"] = y_pred
```

```python
df_prediction["prediction_label"] = df_prediction["prediction"].map({0: "sentosa", 1: "versicolor", 2: "virginica"})
```

```python
def plot_incorrect_prediction(df_prediction, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Adjust the figure size as needed
    
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_prediction, ax=axs[0, 0])
    axs[0, 0].set_title("Prediction Label")
    axs[0, 0].set_xlabel(x_axis_feature)
    axs[0, 0].set_ylabel(y_axis_feature)

    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_prediction, ax=axs[0, 1])
    axs[0, 1].set_title("Target Name")
    axs[0, 1].set_xlabel(x_axis_feature)
    axs[0, 1].set_ylabel(y_axis_feature)

    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_prediction", data=df_prediction, ax=axs[1, 0])
    axs[1, 0].set_title("Correct Prediction")
    axs[1, 0].set_xlabel(x_axis_feature)
    axs[1, 0].set_ylabel(y_axis_feature)
    axs[1, 1].axis('off')

    plt.tight_layout() 
    plt.show()
```

```python
plot_incorrect_prediction(df_prediction, "petal length (cm)", "petal width (cm)")
```
